```python
numeric_cols = ["avg_rating", "rating_count", "popularity_score", "num_genres"]
corr = features_df[numeric_cols].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Numeric Features")
plt.show()
```
